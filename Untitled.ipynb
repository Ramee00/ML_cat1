```python
import pandas as pd
from numpy import random
import matplotlib.pyplot as plt 
```

```python
x = random.randint(1000,size=(1000))
print(x) 
```

```python
y = random.randint(1000,size=(1000))
print(y) 
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=11)
```

```python
x_mean=x_train.mean()
```

```python
y_mean=y_train.mean()
```

```python
l=len(x_train)
```

```python
num=0
den=0
```

```python
for i in range(l):
    num=num+((x_train[i]-x_mean)*(y_train[i]-y_mean))
    den=den+((x_train[i]-x_mean)**2)
m=num/den
```

```python
c=y_mean-(m*x_mean)
```

```python
print(num,den,m,c)
```

```python
#plotting
plt.scatter(x_train,y_train)
plt.plot(x_train,m*x_train+c)
```

```python
y_pred=m*x_test+c
```
