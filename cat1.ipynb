```python
import pandas as pd
from numpy import random
import matplotlib.pyplot as plt 
import math as mt
```

1000 random points (x,y)

```python
x = random.randint(1000,size=(1000))
print(x) 
```

```python
y = random.randint(1000,size=(1000))
print(y) 
```

simple linear regression from scratch

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=11)
```

```python
x_mean=x_train.mean()
```

```python
y_mean=y_train.mean()
```

```python
l=len(x_train)
```

```python
num=0
den=0
```

```python
for i in range(l):
    num=num+((x_train[i]-x_mean)*(y_train[i]-y_mean))
    den=den+((x_train[i]-x_mean)**2)
m=num/den
```

```python
c=y_mean-(m*x_mean)
```

```python
print(num,den,m,c)
```

```python
#plotting
plt.scatter(x_train,y_train)
plt.plot(x_train,m*x_train+c)
```

```python
y_pred=m*x_test+c
```

```python
y_pred
```

```python
e=(y_test-y_pred)
```

```python
e
```

```python
mse=(((e)**2)/l).sum()
```

```python
rmse=mt.sqrt(mse)
```

```python
m1=m
c1=c
lr=0.01
iterations=1000
error=[]
```

Gradient Descent to find the optimal slope and intercept value

```python
for i in range(iterations):
    cost=0
    cost_c=0
    cost_m=0
    y_pred1=(c1+m1*x_test[i])
    cost=cost+(y_test[i]-y_pred1)**2
    partial_c=2*(y_test[i]-(c+m*x_test[i]))
    partial_m=2*(y_test[i]-(c+m*x_test[i]))
    cost_c=cost_c+partial_c
    cost_m=cost_m+partial_m   
    c1=c1-lr*cost_c
    m1=m1-lr*cost_m
        
    error.append(cost)
```
